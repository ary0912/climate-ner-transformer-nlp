```python
import matplotlib.pyplot as plt
import seaborn as sns

models = ["NB Baseline", "NB Improved", "FFNN", "BERT-tiny"]
accuracies = [0.65, 0.79, 0.475, 0.66]

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
bars = sns.barplot(x=models, y=accuracies, palette="Set2")

for bar in bars.patches:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.015, f"{height:.2f}", ha='center', fontsize=12)

plt.title("Validation Accuracy of Models", fontsize=16)
plt.ylabel("Accuracy")
plt.ylim(0.4, 0.85)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

models = ["NB Baseline", "NB Improved", "FFNN", "BERT-tiny"]
accuracies = [0.65, 0.79, 0.475, 0.66]

plt.figure(figsize=(8,5))
bars = plt.bar(models, accuracies, color=['gray', 'skyblue', 'lightgreen', 'salmon'])
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f"{yval:.2f}", ha='center')
plt.ylim(0.4, 0.85)
plt.title("Validation Accuracy of All Models")
plt.ylabel("Accuracy")
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python

epochs = list(range(1, 6))
train_loss = [1.2, 0.9, 0.75, 0.6, 0.58]
val_loss =   [1.3, 1.1, 1.05, 1.1, 1.2]

plt.figure(figsize=(8,5))
plt.plot(epochs, train_loss, marker='o', label="Train Loss")
plt.plot(epochs, val_loss, marker='o', label="Validation Loss")
plt.title("FFNN Learning Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
